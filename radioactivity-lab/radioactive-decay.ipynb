# Exponential Decay Simulation

The goal of this project is to simulate an activity performed in PHYS 4C.
That activity "simulated" radioactive decay by rolling dice.
The dice used were mostly blank, but a few sides had a single dot.
Each dice represented a nucleus, and a dot coming up represented the nucleus randomly decaying.
Each roll represented the passage of time in this model.
In real radioactive decay, unstable nuclei have a random chance of decaying at any time.
Similarly, the dice had a random chance of landing on a side with a spot on any roll.
By using code, the activity can be performed more quickly with easily adjustable parameters.

```python
import math
import random
import matplotlib.pyplot as plt
```

```python
# Dice properties
decay_sides = 2
total_sides = 20
# Starting amount and trial count
N_0 = 140
trial_count = 25
trials = list(range(0, trial_count + 1))
```

```python
# Simulate the decays
N = [N_0]
for t in range(trial_count):
    decay_count = 0
    for __ in range(N[t]):
        num_rolled = random.randint(1, total_sides)
        if num_rolled > decay_sides:
            pass
        else:
            decay_count += 1
    N.append(N[t] - decay_count)

# Calculate the theoretical counts
decay_constant = math.log(1 / (1 - decay_sides / total_sides))
N_th = [N_0 * math.exp(-decay_constant * t) for t in trials]
```

```python
# Create a graph of the simulated and theoretical counts over the course of the trials
fig, ax = plt.subplots()
# Plot the points
ax.scatter(trials, N, label="Simulated")
ax.scatter(trials, N_th, label="Theoretical")
# Label the axes
ax.set_xlabel("Trial")
ax.set_ylabel("Dice Remaining")
# Set a title
ax.set_title("Dice Remaining After Each Trial")
# Add a legend
ax.legend()
```
